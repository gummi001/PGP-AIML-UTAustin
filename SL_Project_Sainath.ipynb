```python
col=list(Bank_combined.select_dtypes(include=['int64','float64']).columns)
```

```python
for i in col:
    x1 = np.array(Bank_combined[Bank_combined.LoanOnCard == 0][i]) 
    y1 = np.array(Bank_combined[Bank_combined.LoanOnCard == 1][i])
    t, p_value  = stats.ttest_ind(x1,y1, axis = 0,equal_var=False) 
    print('{} P_Value:{}'.format('\033[1m',p_value))
    if p_value < 0.05:  # Setting our significance level at 5%
        print('{} Rejecting Null Hypothesis.{} of Loan holders and non-Loan holders are not same'.format('\033[1m',i))
    else:
        print('{} Fail to Reject Null Hypothesis.{} of Loan holders and non-Loan holders are  same'.format('\033[1m',i))
    print('\n')
```

```python
for c in col:
    #getting upper lower quartile values
    q25,q75=np.percentile(Bank_combined[c],25),np.percentile(Bank_combined[c],75)
    IQR=q75-q25
    Threshold=IQR*1.5
    lower,upper=q25-Threshold,q75+Threshold
    Outliers=[i for i in Bank_combined[c] if i < lower or i > upper]
    print('{} Total Number of outliers in {} Before Imputing : {}'.format('\033[1m',c,len(Outliers)))
    print('\n')
    #taking mean of a column without considering outliers
    Bank_combined_include = Bank_combined.loc[(Bank_combined[c] >= lower) & (Bank_combined[c] <= upper)]
    mean=int(Bank_combined_include[c].mean())
    print('{} Mean of {} is {}'.format('\033[1m',c,mean))
    print('\n')
    #imputing outliers with mean
    Bank_combined[c]=np.where(Bank_combined[c]>upper,mean,Bank_combined[c])
    Bank_combined[c]=np.where(Bank_combined[c]<lower,mean,Bank_combined[c])
    Outliers=[i for i in Bank_combined[c] if i < lower or i > upper]
    print('{} Total Number of outliers in {} After Imputing : {}'.format('\033[1m',c,len(Outliers)))  
    print('\n')
```
